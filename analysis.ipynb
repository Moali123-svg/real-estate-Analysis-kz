```python
import pandas as pd
import numpy as np;
import seaborn as sns;
import matplotlib.pyplot as plt
```

```python
CSV_PATH = 'krishakz.csv'
df = pd.read_csv(CSV_PATH)
```

```python
df
```

```python
#Data preprocessing 
df['price'] = df.price.str.replace(' 〒', '', regex = False)
df['price'] = df.price.str.replace(' ', '', regex = False)
df['price'] = df['price'].astype('int')
df['area'] = df.area.str.replace(' помесячно', '', regex = False)
df['area'] = df.area.str.replace(' м²', '', regex = False)
df['area'] = df['area'].astype('float')
```

```python
#delete outliers
df = df[(df.name != '11-комнатная квартира') & (df.distinct != 'Аль Фараби 7') & (df.distinct != 'Аль Фараби 7а')]
```

```python
df.info()
```

```python
#number of apartments for rent in each district
count_by_district = df.groupby(['distinct'], as_index=False)['price'].count().sort_values(['price'],ascending=False)
count_by_district
```

```python
df = df[df.groupby('name').distinct.transform('count')>10].copy()
df.groupby(['name'], sort=False)['name'].count().plot.pie(y='name',figsize=(5, 5),autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
#average price per apartment type in each district 
med_price_by_type = df.groupby(['name','distinct'], as_index=False)['price'].median().sort_values(['price'],ascending=False)
table = pd.pivot_table(med_price_by_type, values='price', index=['distinct'],columns=['name'])

table
```

```python
#heatmap for average price
a = sns.heatmap(table,cmap="Reds")
```

```python
#Correlation of price and area
price = df['price'];
area = df['area'];

price.corr(area)
```

```python
plt.plot(area, price, 'ro')
plt.axis([0, 600, 0, 2600000])
plt.show()
```

```python
#average price in each district
med_price_by_district = df.groupby('distinct', sort=False).agg('median')[['price']].sort_values(['price'],ascending=True) 
med_price_by_district
```

```python
#max and min prices for each apartment type
max_min_price = df.groupby('name', sort=False).agg({'price' : ['max', 'min','median']}).sort_values(['name'],ascending=True) 
max_min_price
```

```python
max_min_price.plot(kind='bar', stacked=False)
```

```python
#max and min areas for each apartment type
max_area = df.groupby('name', sort=False).agg({'area' : ['max', 'min','median']}).sort_values(['name'],ascending=True) 
max_area
```

```python
max_area.plot(kind='bar', stacked=False)
```
